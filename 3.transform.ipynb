```python
from pathlib import Path 
import chardet 
import os 
import pandas as pd
import numpy as np
```

```python
root_dir = Path(".").resolve().parent
root_dir
```

```python
files_path=os.path.join(root_dir,"etl","data","betaAparthments.csv")
```

```python
df = pd.read_csv(files_path,encoding='utf-8') 
df
```

```python
df_provision = df[['provision']]
df_provision_dummies = pd.get_dummies(df_provision)
df_condition = df[['condition']]
df_condition_dummies = pd.get_dummies(df_condition)
df = pd.concat([df,df_provision_dummies, df_condition_dummies], axis=1)
```
